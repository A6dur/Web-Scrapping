```python
import requests

for i in range(1 , 51):
    a = requests.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
    with open(f"html/textfile{i}.html" , "w" , encoding="utf-8") as f:
        f.write(a.text)
        print(f"File {i} downloaded")

```

```python
from bs4 import BeautifulSoup
import requests

items = []
for i in range(1,51):
    with open (f"html/textfile{i}.html", encoding='utf-8') as f:
            content = f.read()
           
        
    soup = BeautifulSoup(content, "html.parser")
    articles = soup.select("article.product_pod")

    for article in articles :
        title = article.find('h3').find('a')['title']
        price = article.select_one("p.price_color").text.split("£")[1]
        items.append([title , price])
            
    
```
